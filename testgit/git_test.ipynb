```python
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(load_iris()["data"], columns = load_iris()["feature_names"])
df["target"] = load_iris()["target"]
df.head()
```

```python
%matplotlib notebook
from matplotlib.lines import Line2D

fig, ax = plt.subplots()

colors = ["#f9a633" if t == 0 else "#645822" if t == 1 else "#f1441d" for t in df["target"]]
ax.scatter(df.iloc[:, 0], df.iloc[:, 1], c = colors, marker = "o")
ax.set_xlabel(df.columns[0])
ax.set_ylabel(df.columns[1])
ax.set_title(f"{df.columns[0]} vs {df.columns[1]}\n Iris Dataset")

c1 = Line2D(range(1), range(1), color="white", marker='o', markerfacecolor= "#f9a633", markersize = 8)
c2 = Line2D(range(1), range(1), color="white", marker='o', markerfacecolor="#645822", markersize = 8)
c3 = Line2D(range(1), range(1), color="white", marker='o', markerfacecolor="#f1441d", markersize = 8)

plt.legend((c1,c2,c3),load_iris()["target_names"],numpoints=1, loc=1)
plt.show()
```
